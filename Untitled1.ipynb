```python
outliers = {}
def detect_outliers_iqr(column):
    o = []
    data = df[column]
    q1 = n.percentile(data, 25)
    q3 = n.percentile(data, 75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    for i in data:
        if i < lower_bound or i > upper_bound:
            o.append(i)

    outliers[column] = len(o)

detect_outliers_iqr('feature_1')
detect_outliers_iqr('feature_2')
detect_outliers_iqr('feature_3')
detect_outliers_iqr('feature_4')
detect_outliers_iqr('feature_5')
detect_outliers_iqr('feature_6')
detect_outliers_iqr('feature_7')
detect_outliers_iqr('feature_8')



print(outliers)
```

```python
# use clipping to remove the outliers
def clip_outliers(column):
    data = df[column]
    q1 = n.percentile(data, 25)
    q3 = n.percentile(data, 75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    data = n.clip(data, lower_bound, upper_bound)
    return data

df['feature_1'] = clip_outliers('feature_1')
df['feature_2'] = clip_outliers('feature_2')
df['feature_3'] = clip_outliers('feature_3')
df['feature_4'] = clip_outliers('feature_4')
df['feature_5'] = clip_outliers('feature_5')
df['feature_6'] = clip_outliers('feature_6')
df['feature_7'] = clip_outliers('feature_7')
df['feature_8'] = clip_outliers('feature_8')

```

```python
from scipy import stats

# List of features to compare against the target
n_features = ["feature_1", "feature_2", "feature_3", "feature_4", "feature_5", "feature_6", "feature_7", "feature_8"]

# Perform t-tests for each feature against the target
for feature in n_features:
    # Select the data for target = 0 and target = 1
    group_0 = df[df["target"] == 0][feature]
    group_1 = df[df["target"] == 1][feature]

    # Perform t-test (Welchs t-test, which does not assume equal variance)
    t_stat, p_value = stats.ttest_ind(group_0, group_1, equal_var=False)

    # Print the results: t-statistic and p-value
    print(f"{feature}: T-statistic = {t_stat:.4f}, P-value = {p_value:.4f}")
```

```python
from sklearn.ensemble import VotingClassifier, RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer # Import SimpleImputer
import pandas as ps

# Load the dataset
df = ps.read_csv('final_project_dataset_complete.csv')

# Separate features (X) and target (y)
X = df.drop('target', axis=1)
y = df['target']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Define a preprocessor to handle categorical data (OneHotEncoder), numeric data (passthrough), and missing values (SimpleImputer)
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', Pipeline(steps=[
            ('imputer', SimpleImputer(strategy='most_frequent')),  # Impute missing values in categorical features
            ('onehot', OneHotEncoder())  # Apply OneHotEncoder to categorical columns
        ]), X.select_dtypes(include=['object']).columns),
        ('num', Pipeline(steps=[
            ('imputer', SimpleImputer(strategy='median')),  # Impute missing values in numerical features
            ('passthrough', 'passthrough')  # Leave numeric columns as they are
        ]), X.select_dtypes(include=['int64', 'float64']).columns)
    ]
)

# Create individual classifiers
log_v = LogisticRegression()
dt_v = DecisionTreeClassifier()
rfc_v = RandomForestClassifier(n_estimators=25, min_samples_split=5, min_samples_leaf=2, max_depth=10)

# Create the VotingClassifier with multiple models
voting_clf = VotingClassifier(estimators=[('lr', log_v), ('dt', dt_v), ('rfc', rfc_v)], voting='hard')

# Create a pipeline with preprocessing and VotingClassifier
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('voting_clf', voting_clf)
])

# Train the VotingClassifier model
pipeline.fit(X_train, y_train)

# Make predictions
voting_pred = pipeline.predict(X_test)

# Assuming model_report is a function for model evaluation
print('Voting Classifier')
model_report(y_test, voting_pred)
```
